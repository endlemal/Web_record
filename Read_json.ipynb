```python
lengths = []
for item in ms_obj['data']:
    try:
        # 尝试获取深度嵌套的值
        detail_list = item['RoomInfo']['data']['mpDetailList']
        lengths.append(len(detail_list))
    except (KeyError, TypeError):
        # 如果路径不存在或类型错误（如 data 是列表），则跳过或记录为 0
        lengths.append(0)

print(lengths)
```

```python
from urllib.parse import unquote


lengths = []
for item in ms_obj['data'][5]['RoomInfo']['data']['mpDetailList']:
    try:

        uname = unquote(item['nickName'])
        lengths.append(uname)
    except (KeyError, TypeError):

        lengths.append(0)

print(lengths)
```
